```python
import json

import numpy as np
import requests


def get_moves(state, n=10):
    url = "http://liacpc17.epfl.ch:5001/legal_moves"
    o = {"smiles": state}
    response = requests.post(url, json=o).json()["smiles_list"]
    # Select 10 at random
    try:
        return np.random.choice(
            response, min(n, len(response)), replace=False
        ).tolist()
    except:
        return response


def get_task(task, n=10):
    return {
        "id": task[0] + ">>" + task[-1],
        "rxn": task[0] + ">>" + task[-1],
        "steps": task,
        "step_options": [
            get_moves(task[i], n=n) for i in range(0, len(task) - 1)
        ],
    }


ftasks = [get_task(task, n=6) for task in tasks]
with open("benchmark.json", "w") as f:
    json.dump(ftasks, f, indent=2)
```
