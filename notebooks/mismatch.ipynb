```python
import os
import json
import numpy as np
from steer.evaluation.synthesis import load_default_tasks

def get_latest_file(path, fid):
    """Get the most recent file for a given task id."""
    import os

    files = os.listdir(path)
    files = [f for f in files if fid in f]
    files.sort()
    if files:
        return f"{path}/{files[-1]}"
    else:
        return None


data_dir = '../data/outputs/2025-06-14_125025'  # Results from AIZynth
tasks = load_default_tasks('../data/synthia_routes')
thr = 7

c = 0
miss_routes = []
for t in tasks:
    route = get_latest_file(data_dir, t.id)
    with open(route, "r") as f:
        routes = json.load(f)
    gt_score, lmscore = t.evaluate(routes)

    # If any difference is higher than t, print the task id and scores
    idx = abs(np.array(gt_score) - np.array(lmscore)) > thr
    for r in np.where(idx)[0]:
        print(f"Example #{c}. Task {t.id} - Route {r}: GT Score = {gt_score[r]}, LM Score = {lmscore[r]}")
        print("-" * 40)
        miss_routes.append(routes[r])
        c+= 1

with open("missmatch_routes.json", "w") as f:
    json.dump(miss_routes, f, indent=4)
```
